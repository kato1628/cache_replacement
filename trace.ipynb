```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


trace = pd.read_csv(
    './dataset/wiki2018_sample.tr',
    delim_whitespace=True,
    names=["time", "id", "size", "type"]
)
sizes = trace["size"].to_numpy()
min_size, max_size = min(sizes), max(sizes)

bin_min, bin_max = 0, 40000
sizes = np.clip(sizes, bin_min, bin_max)
num_bins = 20

plt.hist(
    sizes,
    color='blue',
    edgecolor='black',
    bins=num_bins,
    range=(bin_min, bin_max)
)
plt.xlabel(f"Size | min: {min_size} max: {max_size}")
plt.ylabel("Frequency")
plt.title("Histogram of Data Size")
plt.show()
```
